```python

# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Set style
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)

print("Libraries loaded!")
```

```python
# Check data types and missing values
print("DATA TYPES:")
print(df.dtypes)
print("\n" + "="*50)

print("\nMISSING VALUES:")
missing = df.isnull().sum()
missing_pct = (missing / len(df)) * 100
missing_table = pd.DataFrame({'Missing': missing, 'Percentage': missing_pct.round(1)})
print(missing_table[missing_table['Missing'] > 0])
```

```python
# Clean data
df_clean = df.copy()

# Fill missing "Age" with median
age_median = df_clean['Age'].median()
missing_age = df_clean['Age'].isnull().sum()
df_clean['Age'] = df_clean['Age'].fillna(age_median)
print(f"Filled {missing_age} missing ages with median ({age_median:.1f})")

# Fill missing "Embarked" with mode
embarked_mode = df_clean['Embarked'].mode()[0]
missing_embarked = df_clean['Embarked'].isnull().sum()
df_clean['Embarked'] = df_clean['Embarked'].fillna(embarked_mode)
print(f"Filled {missing_embarked} missing embarkment")

# Drop Cabin
cabin_missing = df_clean['Cabin'].isnull().sum()
df_clean = df_clean.drop('Cabin', axis=1)
print(f"Dropped Cabin ({cabin_missing} missing, {cabin_missing/len(df_clean)*100:.1f}%)")

# Create age groups
bins = [0, 12, 18, 35, 60, 100]
labels = ['Child (0-12)', 'Teen (13-18)', 'Young Adult (19-35)',
          'Adult (36-60)', 'Senior (60+)']
df_clean['AgeGroup'] = pd.cut(df_clean['Age'], bins=bins, labels=labels)

print("Created age groups")
print("\nData cleaning complete!")
```

```python
# Calculate survival rates
overall = df_clean['Survived'].mean() * 100
print(f"OVERALL SURVIVAL RATE: {overall:.1f}%")
print(f"({df_clean['Survived'].sum()} survivors out of {len(df_clean)})\n")

#SURVIVAL BY SEX
print("-" * 40)
print("SURVIVAL BY SEX:")
sex_surv = df_clean.groupby('Sex')['Survived'].mean() * 100
for sex, rate in sex_surv.items():
    count = df_clean[df_clean['Sex'] == sex].shape[0]
    print(f"  {sex:6s}: {rate:5.1f}%  ({count} passengers)")

#SURVIVAL BY PASSENGER CLASS
print("\n" + "-" * 40)
print("SURVIVAL BY PASSENGER CLASS:")
class_surv = df_clean.groupby('Pclass')['Survived'].mean() * 100
for pclass, rate in class_surv.items():
    count = df_clean[df_clean['Pclass'] == pclass].shape[0]
    print(f"  Class {pclass}: {rate:5.1f}%  ({count} passengers)")

#SURVIVAL BY AGE GROUP
print("\n" + "-" * 40)
print("SURVIVAL BY AGE GROUP:")
age_surv = df_clean.groupby('AgeGroup', observed=False)['Survived'].mean() * 100
for age_group, rate in age_surv.items():
    count = df_clean[df_clean['AgeGroup'] == age_group].shape[0]
    print(f"  {age_group:20s}: {rate:5.1f}%  ({count} passengers)")
```

```python
# Imputing missing values for age
df.fillna({'Age': df['Age'].mean()}, inplace=True)
#Imputing missing values for embarked
df.fillna({'Embarked': 'S'}, inplace=True)


df['SibSp'].value_counts()
df['Parch'].value_counts()
```

```python
#Changing data type
df['Survived']=df['Survived'].astype('category')
df['Pclass']=df['Pclass'].astype('category')
df['Sex']=df['Sex'].astype('category')
df['Age']=df['Age'].astype('int')
df['Embarked']=df['Embarked'].astype('category')
df.info()
```

```python
# Age Distribution by Passenger class and Survival Status
palette = {0: 'red', 1: 'green'}
plt.figure(figsize=(8, 5))
sns.boxplot(data=df_clean, x='Pclass', y='Age', hue='Survived', palette=palette)
plt.title('Age Distribution by Passenger Class and Survival Status',fontsize=15, fontweight='bold')
plt.xlabel('Passenger Class', fontsize=12)
plt.ylabel('Age', fontsize=12)
handles, _ = plt.gca().get_legend_handles_labels()
labels = ['No (Died)', 'Yes (Survived)']
plt.legend(handles, labels, title='Survived')
plt.tight_layout()
plt.savefig('outputs/age_class_boxplot.png', dpi=150, bbox_inches='tight')
plt.show()
print("Saved: outputs/age_class_boxplot.png")
```

```python
# Correlation heatmap
plt.figure(figsize=(8, 6))
numeric_cols = df_clean.select_dtypes(include=[np.number]).columns
corr = df_clean[numeric_cols].corr()
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, mask=mask, annot=True, cmap='coolwarm', center=0,
            square=True, linewidths=1, fmt='.2f')
plt.title('Feature Correlations', fontsize=15, fontweight='bold')
plt.tight_layout()
plt.savefig('outputs/correlation_heatmap.png', dpi=150, bbox_inches='tight')
plt.show()
print("Saved: outputs/correlation_heatmap.png")
```

```python
# Create visualizations (Dashboard)
os.makedirs('outputs', exist_ok=True)

fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Plot 1: Survival by Sex
sns.barplot(data=df_clean, x='Sex', y='Survived', hue='Sex', ax=axes[0,0], palette='pastel', legend=False)
axes[0,0].set_title('Survival Rate by Gender', fontsize=14, fontweight='bold')
axes[0,0].set_ylim(0, 1)
for p in axes[0,0].patches:
    axes[0,0].text(p.get_x() + p.get_width()/2., p.get_height() + 0.02,
                   f'{p.get_height()*100:.0f}%', ha='center', fontsize=12)

# Plot 2: Survival by Class
sns.barplot(data=df_clean, x='Pclass', y='Survived', hue='Pclass', ax=axes[0,1], palette='Set2', legend=False)
axes[0,1].set_title('Survival Rate by Passenger Class', fontsize=14, fontweight='bold')
axes[0,1].set_ylim(0, 1)
for p in axes[0,1].patches:
    axes[0,1].text(p.get_x() + p.get_width()/2., p.get_height() + 0.02,
                   f'{p.get_height()*100:.0f}%', ha='center', fontsize=12)

# Plot 3: Age distribution
sns.violinplot(data=df_clean, x='Survived', y='Age', hue='Survived',
               palette='muted', legend=False, ax=axes[1,0])
axes[1,0].set_title('Age Distribution: Survivors vs Non-Survivors', fontsize=14, fontweight='bold')
axes[1,0].set_xticks([0, 1])
axes[1,0].set_xticklabels(['Died', 'Survived'])

# Plot 4: Survival by Age Group
age_surv.plot(kind='bar', ax=axes[1,1], color='skyblue', edgecolor='navy', linewidth=2)
axes[1,1].set_title('Survival Rate by Age Group', fontsize=14, fontweight='bold')
axes[1,1].set_ylabel('Survival Rate (%)')
axes[1,1].set_ylim(0, 100)
axes[1,1].axhline(y=overall, color='red', linestyle='--', linewidth=2, label=f'Overall: {overall:.1f}%')
axes[1,1].legend()
axes[1,1].set_xticklabels(age_surv.index, rotation=45, ha='right')

plt.suptitle('Titanic Survival Analysis', fontsize=18, fontweight='bold', y=1.02)
plt.tight_layout()
plt.savefig('outputs/titanic_analysis.png', dpi=150, bbox_inches='tight')
plt.show()
print("Saved: outputs/titanic_analysis.png")
```
